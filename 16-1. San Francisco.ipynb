```python
import csv

from matplotlib import pyplot as plt

# Get high temperatures from file
filename = "san_francisco_weather.csv"
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    highs = []
    for row in reader:
        high = int(row[11])
        highs.append(high)
    
# Plot data.
fig = plt.figure(dpi=128, figsize=(8,4))
plt.plot(highs, c='red')

# Format plot.
plt.title("San Francisco daily high temperatures, 2014", fontsize = 24)
plt.xlabel('', fontsize=16)
plt.ylabel("Temperature (F)", fontsize=16)
plt.tick_params(axis="both", which="major", labelsize=16)

plt.show()
```

```python
import csv
from datetime import datetime

from matplotlib import pyplot as plt

# Get dates, high temperatures, and low temperatures from file.
filename = "san_francisco_weather.csv"
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    dates, highs, lows = [], [], []
    for row in reader:
        current_date = datetime.strptime(row[5], "%Y-%m-%d")
        dates.append(current_date)
        
        high = int(row[11])
        highs.append(high)
        
        low = int(row[12])
        lows.append(low)
    
# Plot data.
fig = plt.figure(dpi=128, figsize=(8,4))
plt.plot(dates, highs, c="red")
plt.plot(dates, lows, c="blue")

# Format plot.
plt.title("San Francisco daily high and low temperatures, 2014", fontsize = 24)
plt.xlabel('', fontsize=16)
plt.ylabel("Temperature (F)", fontsize=16)
plt.tick_params(axis="both", which="major", labelsize=16)

fig.autofmt_xdate()
plt.ylabel("Temperature (F)", fontsize=16)
plt.tick_params(axis="both", which="major", labelsize=16)

plt.show()
```

```python
import csv
from datetime import datetime

from matplotlib import pyplot as plt

# Get dates, high temperatures, and low temperatures from file.
filename = "san_francisco_weather.csv"
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    dates, highs, lows = [], [], []
    for row in reader:
        current_date = datetime.strptime(row[5], "%Y-%m-%d")
        dates.append(current_date)
        
        high = int(row[11])
        highs.append(high)
        
        low = int(row[12])
        lows.append(low)
    
# Plot data.
fig = plt.figure(dpi=128, figsize=(8,4))
plt.plot(dates, highs, c="red", alpha=0.5)
plt.plot(dates, lows, c="blue", alpha=0.5)
plt.fill_between(dates, highs, lows, facecolor="blue", alpha=0.1)

# Format plot.
plt.title("San Francisco daily high and low temperatures, 2014", fontsize = 24)
plt.xlabel('', fontsize=16)
plt.ylabel("Temperature (F)", fontsize=16)
plt.tick_params(axis="both", which="major", labelsize=16)

fig.autofmt_xdate()
plt.ylabel("Temperature (F)", fontsize=16)
plt.tick_params(axis="both", which="major", labelsize=16)

plt.show()
```
