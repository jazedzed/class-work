```python
df = pd.read_csv("7-5_weather_data.csv", parse_dates=["day"])
df.set_index("day", inplace=True)
df
```

```python
new_df = df.fillna(0)
new_df
```

The syntax above creates a new dataframe and fills NaN values with "0."

```python
new_df = df.fillna({
    "temperature": 0,
    "windspeed": 0,
    "event": "no event"
})
new_df
```

The above syntax allows you to guess what the value of a column might be. For instance, in event, we can enter "no event," which is more meaningful than "0."

```python
new_df = df.fillna(method="ffill")
new_df
```

The syntax above, "ffill," lets you carry forward the value from the previous day.

```python
new_df = df.fillna(method="bfill")
new_df
```

The syntax above, "bfill," allows you to backfill data from the following day.

```python
new_df = df.fillna(method="bfill", axis="columns")
new_df
```

The syntax above lets you copy values horizontally.

```python
new_df = df.fillna(method="ffill", limit=1)
new_df
```

The syntax above, limit=1, is supposed to let you specify a limit on the number of cells you want to carry a value forward. However, it doesn't appear to be working for me.

```python
new_df = df.interpolate()
new_df
```

The .interpolate() method (above) returns a value for NaN values that is between the existing known values. For instance, on 1/4, the temp is guessed at 30, which is between 32 and 28.

```python
new_df = df.interpolate(method="time")
new_df
```

By specifying method="time," you can ask the program to interpolate a value that is dependent on the date. So the temp for 1/4 is now 29, because 1/4 is closer to 1/5 than to 1/1.

```python
new_df = df.dropna()
new_df
```

The .dropna() method allows you to drop all rows with NaN values.

```python
new_df = df.dropna(how="all")
new_df
```

By specifying how="all," you can ask the program to drop only those rows with ALL NaN values.

```python
new_df = df.dropna(thresh=2)
new_df
```

By specifying thresh=2, you can ask the program to keep all rows with at least two valid non NaN values.

```python
dt = pd.date_range("01/01/17", "01/11/17")
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```

The above syntax allows you to create rows for dates that were missing from the original dataframe.
