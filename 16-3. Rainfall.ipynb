```python
import csv
from datetime import datetime

from matplotlib import pyplot as plt

# Get precipitation from file.
filename = "baltimore_weather.csv"
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    dates, precipitation = [], []
    for row in reader:
        current_date = datetime.strptime(row[5], "%Y-%m-%d")
        dates.append(current_date)

        precip = float(row[7])
        precipitation.append(precip)
    
# Plot data.
fig = plt.figure(dpi=128, figsize=(8,4))
plt.plot(dates, precipitation, c="blue")

# Format plot.
plt.title("Baltimore daily precipitation, 2014", fontsize = 24)
plt.xlabel('', fontsize=16)
plt.ylabel("Precipitation (inches)", fontsize=16)
plt.tick_params(axis="both", which="major", labelsize=16)

fig.autofmt_xdate()
plt.ylabel("Precipitation (inches)", fontsize=16)
plt.tick_params(axis="both", which="major", labelsize=16)

plt.show()
```
