### 15-1. Cubes

```python
import matplotlib.pyplot as plt

input_values = [1, 2, 3, 4, 5]
cubes = [1, 8, 27, 64, 125]
plt.plot(input_values, cubes, linewidth=5)

# Set chart title and label axes.
plt.title("Cubic Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Cube of Value", fontsize=14)

# Set size of tick labels.
plt.tick_params(axis="both", which="major", labelsize=14)

plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = list(range(1, 50001))
y_values = [x**3 for x in x_values]

plt.scatter(x_values, y_values, s=40)

# Set chart title and label axes.
plt.title("Cubic Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Cube of Value", fontsize=14)

# Set the range for each axis.
plt.axis([0, 5100, 0, 126000000000])

plt.show()
```

### 15-2. Colored Cubes

```python
import matplotlib.pyplot as plt

x_values = list(range(1, 50001))
y_values = [x**3 for x in x_values]

plt.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Reds, edgecolor="none", s=40)

# Set chart title and label axes.
plt.title("Cubic Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Cube of Value", fontsize=14)

# Set the range for each axis.
plt.axis([0, 5100, 0, 126000000000])

plt.show()
```
