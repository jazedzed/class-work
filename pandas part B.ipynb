### This dataset represents English-language fiction published between 1700 and 1922, and contained in the HathiTrust Digital Library.

```python
import pandas as pd
df = pd.read_csv("fiction_metadata.csv")
df
```

### Question 1: What is the title published between 1700 and 1922 containing the greatest number of pages?

```python
df[df.totalpages==df.totalpages.max()]
```

### Answer 1: Rolland Romain's *Jean Christophe* contains the greatest number of pages of the novels in this dataset, at 1,600 total pages. The two records represent different years in which the novel was published.

### Question 2: What were all of the titles published by Nathaniel Hawthorne?

```python
g = df.groupby("author")
g
```

```python
g.get_group("Hawthorne, Nathaniel,")
```

### Answer 2: All of the books published by Nathaniel Hawthorne appear above.
